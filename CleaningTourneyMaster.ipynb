```python
with connection, connection.cursor() as cursor: 
    
    cursor.execute(
    ''' 
    CREATE TABLE IF NOT EXISTS tournament_table(
        tournament_id  INT NOT NULL, 
        player_id INT NOT NULL,
        purse FLOAT,
        year VARCHAR(10),
        score INT,
        par INT,
        finish VARCHAR(20),
        rounds INT,
        cut INT,
        shots_gained_putt FLOAT,
        shots_gained_around_green FLOAT,
        shots_gained_approach_green FLOAT,
        shots_gained_off_the_tee FLOAT,
        shots_gained_tee_to_green FLOAT,
        shots_gained_total FLOAT,
        top10 INT,
        FOREIGN KEY(tournament_id) REFERENCES tourney_dates(tournament_id),
        FOREIGN KEY(player_id) REFERENCES player_table(player_id)
        
        
    );
    ''')
```

```python
with connection, connection.cursor() as cursor: 
    cursor.execute("SELECT COLUMN_NAME FROM information_schema.COLUMNS WHERE TABLE_NAME ='tournament_table' ;" )
    results=cursor.fetchall()
    
print('Columns in table tournament_table:')
for row in results: 
    print(row)
```
